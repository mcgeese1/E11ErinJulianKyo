```python
import numpy as np
import pandas as pd 
import csv
import matplotlib.pyplot as plt
```

```python
outside_5 = 'Weather_AQI_outside5.csv'
outside_6 = 'outsideweek6.csv'

dfout5 = pd.read_csv(outside_5)
dfout6 = pd.read_csv(outside_6)

dfout5= dfout5[1:]
dfout6= dfout6[1:]
```

```python
plt.scatter(dfout5["Temp"],dfout5["Humidity"],label="Week 5 Data",marker='.') 
plt.scatter(dfout6["Temp"],dfout6["Humidity"] ,label="Week 6 Data",marker='.')
plt.legend()
plt.title("Temperature vs Humidity")
plt.xlabel("Temperature")
plt.ylabel("Humidity")
[m,b]=np.polyfit(dfout5["Temp"],dfout5["Humidity"],deg=1)
[m2,b2]=np.polyfit(dfout6["Temp"],dfout6["Humidity"],deg=1)


xvec = np.linspace(18,24,100)
yvec = xvec*m+b
yvec2 = xvec*m2+b2
plt.plot(xvec,yvec)
plt.plot(xvec,yvec2)
print("Pearson correlation:", np.corrcoef(dfout5["Temp"],dfout5["Humidity"])[0,1])
print("this confirms negative linear correlation for Week 5")

print("Pearson correlation:", np.corrcoef(dfout6["Temp"],dfout6["Humidity"])[0,1])
print("this confirms negative linear correlation for Week 6")

      
plt.show()


```

Yes, there seems to be a strong correlation between humidity and temperature. In both cases,as temperature increases, overall humidity decreases.

```python
x = dfout6['Temp']
y = dfout6['PM 2.5']
plt.plot(x,y)

plt.xlabel('Temperature[C]')
plt.ylabel('PM 2.5 Counts')
plt.title('Temperature vs PM 2.5 Counts Week 6')
plt.show() 

correlation_matrix = np.corrcoef(x, y)
pearson_corr_numpy = correlation_matrix[0, 1]
#print(f"Pearson correlation using NumPy: {pearson_corr_numpy}")

print("Pearson's Correlation number is {:.2f}, quite close to 0, meaning they are not well correlated.".format(pearson_corr_numpy))
```

```python
plt.plot(dfout6["Temp"],dfout6["Pressure"],label="Temperature vs Pressure")
#plt.plot(usefulnew["Pressure"],label="Pressure")
plt.xlabel("Temperature")
plt.title("Temperature vs Pressure")
plt.ylabel("Pressure (hPa)")
plt.legend()
plt.show()

correlation_matrix = np.corrcoef(dfout6["Temp"],dfout6["Pressure"])
print("Pearson's Corrlation number is {:.2f}".format(correlation_matrix[0,1]))
print("so strong negative correlation")
```
