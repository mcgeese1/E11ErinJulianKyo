```python
import numpy as np
import pandas as pd 
import csv
from pathlib import Path
import matplotlib.pyplot as plt

home_dir = Path.home()
#print(home_dir)
```

```python
inside_file = 'Weather_AQI_inside5.csv'
outside_file = 'Weather_AQI_outside5.csv'
their_outside_file = 'others_outsidedata.csv' 
outside_2_file = 'outsideweek6.csv'
```

```python
dfinside = pd.read_csv(inside_file)
dfoutside = pd.read_csv(outside_file)
their_outside = pd.read_csv(their_outside_file)
#display(dfinside)
useful_inside = dfinside[1:]
useful_outside = dfoutside[1:]
useful_their_outside = their_outside[1:]
display(useful_outside)
display(useful_their_outside)
week6_outside = pd.read_csv(outside_2_file)
useful_week6 = week6_outside[1:]
display(useful_week6)
```

PM 1.0:

```python
plt.hist(useful_week6['PM 1.0'],bins = 3,alpha=0.5)
plt.hist(useful_outside['PM 1.0'],bins=3,alpha=0.5)
plt.xlabel("counts per second")
plt.ylabel('frequency')
plt.title('PM 1.0 counts')
plt.show()
```

```python
outside_PM1_mean = np.mean(useful_outside['PM 1.0'])
week6_outside_PM1_mean = np.mean(useful_week6['PM 1.0'])
outside_PM1_std = np.std(useful_outside['PM 1.0'])
week6_outside_PM1_std = np.std(useful_week6['PM 1.0'])
print('a)\n')
print('Our mean uncertainty is {:.4f}'.format(outside_PM1_std/np.sqrt(len(useful_outside['PM 1.0']))))
print('week6 mean uncertainty is {:.4f}'.format(week6_outside_PM1_std/np.sqrt(len(useful_week6['PM 1.0']))))
print('\nb)\n')
print('The means are separated by {:.2f} of our standard deviation'.format(np.abs((outside_PM1_mean-week6_outside_PM1_mean)/(outside_PM1_std/np.sqrt(len(useful_outside['PM 1.0']))))))
print('\nc)\n')
print('There are not enough data points for a gaussian')
```
