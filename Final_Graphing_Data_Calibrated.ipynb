```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Unknown_1_File = 'https://raw.githubusercontent.com/mcgeese1/E11ErinJulianKyo/refs/heads/main/Spectrum_Final_Unknown1_2025-04-18_18-34-46.csv'
Unknown_1_df = pd.read_csv(Unknown_1_File)
Unknown_2_File = 'https://raw.githubusercontent.com/mcgeese1/E11ErinJulianKyo/refs/heads/main/Spectrum_Final_Unknown2_2025-04-18_18-58-28.csv'
Unknown_2_df = pd.read_csv(Unknown_2_File)
Uranium_File = 'https://raw.githubusercontent.com/mcgeese1/E11ErinJulianKyo/refs/heads/main/Spectrum_Final_Uranium_2025-04-18_17-15-40.csv'
Uranium_df = pd.read_csv(Uranium_File)
Thorium_File = 'https://raw.githubusercontent.com/mcgeese1/E11ErinJulianKyo/refs/heads/main/Spectrum_Final_Thorium_2025-04-18_17-41-15.csv'
Thorium_df = pd.read_csv(Thorium_File)
Background_File = 'https://raw.githubusercontent.com/mcgeese1/E11ErinJulianKyo/refs/heads/main/Spectrum_Final_Background_2025-04-18_18-03-15.csv'
Background_df = pd.read_csv(Background_File)
Background_df
```

```python
Total_UK_1 = Unknown_1_df.iloc[[-1]].drop(['Timestamp'],axis=1)
Total_UK_2 = Unknown_2_df.iloc[[-1]].drop(['Timestamp'],axis=1)
Total_U = Uranium_df.iloc[[-1]].drop(['Timestamp'],axis=1)
Total_Th = Thorium_df.iloc[[-1]].drop(['Timestamp'],axis=1)
Total_Background = Background_df.iloc[[-1]].drop(['Timestamp'],axis=1)
Total_UK_1
x_space_uncalibrated = np.linspace(0,1023,1024)
linear_offset = 13.35
linear = 2.23
quadratic = 5e-4
x_space = linear_offset+linear*x_space_uncalibrated+quadratic*x_space_uncalibrated**2
```

```python
plt.plot(x_space, Total_Background.T)
plt.title('Total Counts by Energy for Background')
plt.xlabel('Energy (keV)')
plt.ylabel('Total Counts')
plt.show()
```

```python
plt.plot(x_space, Total_U.T)
plt.title('Total Counts by Energy for Uranium')
plt.xlabel('Energy (keV)')
plt.ylabel('Total Counts')
plt.show()
```

```python
plt.plot(x_space, Total_Th.T)
plt.title('Total Counts by Energy for Thorium 1')
plt.xlabel('Energy (keV)')
plt.ylabel('Total Counts')
plt.show()
```

```python
plt.plot(x_space, Total_UK_1.T)
plt.title('Total Counts by Energy for Unknown Source 1')
plt.xlabel('Energy (keV)')
plt.ylabel('Total Counts')
plt.show()
```
