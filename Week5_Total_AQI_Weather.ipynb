```python
import matplotlib.pyplot as plt
import csv
import numpy as np
import pandas as pd
```

```python
dfinside = pd.read_csv('Weather_AQI_inside5.csv')
dfoutside = pd.read_csv('Weather_AQI_outside5.csv')
useful_inside = dfinside[1:]
useful_outside = dfoutside[1:]
display(dfinside)
display(dfoutside)
```

```python
plt.hist(dfoutside["PM 2.5"],bins=4,label="Outside")
plt.hist(dfinside["PM 2.5"],bins=2,label="Inside")
plt.xlabel("Counts/Sec")
plt.title("PM 2.5 Microns Counts") 
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

Visually, the Outside data has more deviation, however they do not otherwise look significantly different

```python
inside_temp_std = np.std(useful_inside['PM 2.5'])/np.sqrt(300)
inside_temp_mean = np.mean(useful_inside['PM 2.5'])

outside_temp_std = np.std(useful_outside['PM 2.5'])/np.sqrt(300)
outside_temp_mean = np.mean(useful_outside['PM 2.5'])

if outside_temp_mean + 3*outside_temp_std < inside_temp_mean - 3*inside_temp_std:
    print('Number of particles with 2.5 micrometer in length are significantly different!')
else:
    print('Number of particles with 2.5 micrometer in length are not significantly different.')
```

Visually, the Outside data seems to have more counts per second, but it isn't clear if it is statistically different.
