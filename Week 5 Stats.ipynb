```python
import numpy as np
import pandas as pd 
import csv
```

```python
inside_file = 'Weather_AQI_inside5.csv'
outside_file = 'Weather_AQI_outside5.csv'
```

```python
dfinside = pd.read_csv(inside_file)
dfoutside = pd.read_csv(outside_file)
#display(dfinside)
useful_inside = dfinside[1:]
useful_outside = dfoutside[1:]
display(useful_inside)
```

PM 1.0:

```python
inside_temp_std = np.std(useful_inside['PM 1.0'])/np.sqrt(300)
inside_temp_mean = np.mean(useful_inside['PM 1.0'])

outside_temp_std = np.std(useful_outside['PM 1.0'])/np.sqrt(300)
outside_temp_mean = np.mean(useful_outside['PM 1.0'])

if outside_temp_mean + 3*outside_temp_std < inside_temp_mean - 3*inside_temp_std:
    print('Number of particles with 1 micrometer in length are significantly different!')
else:
    print('Number of particles with 1 micrometer in length are not significantly different.')
```
